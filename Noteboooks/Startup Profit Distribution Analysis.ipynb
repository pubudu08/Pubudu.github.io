```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('datasets/MultipleRegression/50_Startups.csv')
x = dataset.iloc[:, :-1].values
print(x)
```

```python
y = dataset.iloc[:, 4].values
print(y)
```

```python
#Encoding categorical variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
#creating the dummy varibales  
x[:,3] = label_encoder.fit_transform(x[:,3])
print(x)
```

```python
one_hot_encoder = OneHotEncoder(categorical_features=[3])
x = one_hot_encoder.fit_transform(x).toarray()
print(x)
```

```python
# Avoiding dummy variable trap
x = x[:, 1:]  # have to remove one dummy variable away, you don't need all three states
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Fit multiple Linear Regression to the training set
from sklearn.linear_model import LinearRegression
# ignore this
import warnings
warnings.filterwarnings('ignore')

regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
# Predicting the Test set result
y_predict = regression.predict(X_test)
```

```python
print("Real profits: ")
print(y_test)
print()
print("Predicted Profits by the model")
print(y_predict)
```

```python
# Building the optimal model using Backward Elimination
import statsmodels.formula.api as sm
```

```python
x = np.append(arr=np.ones((50, 1)).astype(int), values=x, axis=1)
x_opt = x[:, [0, 1, 2, 3, 4, 5]]
```

```python
# only contains independent variables that are highly significant variables that have impact on determining the
# to determine the profit
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
print(regressor_OLS.summary())
```

```python
# removing independent variable that has highest P value which are columns 1,2,4,5
x_opt = x[:, [0, 3]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
print(regressor_OLS.summary())
# This model can be used to forecast effects or impacts of changes.  
# That is, multiple linear regression analysis helps me to understand how much will the dependent variable change 
# when we change the independent variables. 
```
